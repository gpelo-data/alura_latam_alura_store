El **objetivo** de este an√°lisis es determinar cu√°l es la tienda menos eficiente de las 4 tiendas de la cadena AluraStore para recomendar al se√±or Juan que la venda y as√≠ poder comenzar su nuevo emprendimiento.

Para ello se tendr√° en cuenta los siguientes criterios y la tienda que tenga un rendimiento bajo en la mayor√≠a de ellos ser√° la recomendada a vender:  

**An√°lisis de Facturaci√≥n**:

**Valoraci√≥n media por tienda**:  
- Calificaci√≥n promedio

**Rentabilidad**:  
- Valor promedio costo de env√≠o
- Margen aproximado por venta

# üìö Importaci√≥n de librer√≠as

```python
# Para manipular datos
import pandas as pd
import numpy as np
```

```python
# Para manejar ficheros
import os
```

```python
# Visualizaci√≥n de datos
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")     
sns.set_palette("Set2")       
```

```python
# Visualizaci√≥n geogr√°fica
import plotly.express as px
```

# üìÇ Importaci√≥n de datos

```python
# Configuraci√≥n de rutas
file_path = os.getcwd()
file_name = 'tienda'
folder_name = 'datos'
extension = '.csv'

# Lista que guardar√° los dfs
lista_dfs = []

# Cargamos los dfs
for i in range(1,5): # de 1 a 4
    path_base = os.path.join(file_path, folder_name, file_name + str(i) + extension)
    lista_dfs.append(pd.read_csv(path_base))
```

# üîç Exploraci√≥n de los datos

Las 4 tiendas tienen la misma estructura por lo que podemos unifircarlas en un solo df

```python
# Agregar columna identificadora a cada df
lista_dfs[0]['Tienda'] = 'Tienda_1'
lista_dfs[1]['Tienda'] = 'Tienda_2'
lista_dfs[2]['Tienda'] = 'Tienda_3'
lista_dfs[3]['Tienda'] = 'Tienda_4'
```

```python
# Concatenando los df
df = pd.concat(lista_dfs, ignore_index=True)
```

```python
# Resultado
df.head()
```

```python
# Informaci√≥n sobre tipo de dato y valores nulos
df.info()
```

**Observaci√≥n**  
De antemano se sabe que no hay valores nulos o algo que limpiar.

# ‚ôªÔ∏è Funciones

```python
# FUNCI√ìN PARA FORMATEAR UNA CIFRA EN FORMATO ARGENTINO

def formato_arg(numero):
    return f'{numero:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.')
```

```python
# FUNCI√ìN PARA MOSTRAR EL RESULTADO

def print_reporte(df, titulo, tienda, objetivo, formato = False):
    print("\n" + "="*60)
    print(f"{titulo}".center(60))
    print("="*60 + "\n")
    
    for index, row in df.iterrows():
        nombre = '‚óæ ' + row[tienda]
        resultado = formato_arg(row[objetivo]) if formato else row[objetivo]
        print(f"{nombre:15} ‚Üí {resultado:>20}")

    print("\n" + "="*60)
```

# üßÆ An√°lisis de facturaci√≥n

```python
# Agrupamos por tienda y sumamos los valores de la columna Precio
df_facturacion = df.groupby(by='Tienda')['Precio'].sum().reset_index()
```

```python
# Resultado
print_reporte(df_facturacion, 'VENTAS TOTALES', 'Tienda', 'Precio', True)
```

```python
# Gr√°ficamente

sns.barplot(
    data = df_facturacion,
    y = 'Tienda',
    x = 'Precio'
)

plt.title('Comparaci√≥n de volumen de ventas totales')
plt.show()
```

```python
# Contamos la cantidad de transacciones por tienda
df_transacciones = df['Tienda'].value_counts().reset_index()
```

```python
# Resultado
print_reporte(df_transacciones, 'CANTIDAD DE TRANSACCIONES', 'Tienda', 'count')
```

‚≠ï **Interpretaci√≥n**  
La *Tienda_4* presenta una menor cantidad de transacciones (una menos que el resto) y adem√°s es la menor en volumen de ventas.

# üí∏ Ventas por categor√≠a y diversificaci√≥n

```python
# Agrupamos por categor√≠a y sumamos los productos
df_categoria = pd.crosstab(df['Tienda'], df['Categor√≠a del Producto'])
```

```python
# Resultado gr√°fico
plt.figure(figsize=(10,5))
sns.heatmap(
    data = df_categoria,
    annot = True,
    cmap="Set2",
    fmt = '.0f'
)

plt.title('Cantidad de ventas por categor√≠a del producto en cada tienda')
plt.show()
```

‚≠ï **Interpretaci√≥n**  
En **Art√≠culos para el hogar**, Tienda_1 tiene menos ventas.  
En **Deportes y diversi√≥n**, Tienda_2 tiene menos ventas.  
En **Electrodom√©sticos**, Tienda_4 tiene menos ventas.  
En **Electr√≥nicos**, Tienda_2 tiene menos ventas.  
En **Instrumentos musicales**, Tienda_4 tiene menos ventas.  
En **Juguetes**, Tienda_2 tiene menos ventas.  
En **Libros**, Tienda_1 tiene menos ventas.  
En **Muebles**, Tienda_2 tiene menos ventas.  

A groso modo podr√≠amos decir que las tiendas 1, 2 y 4 presentan bajas ventas en al menos una categor√≠a. M√°s adelante veremos si es determinante o no.

```python
# Distribuci√≥n del precio y cuotas

sns.relplot(
    data = df,
    x = 'Cantidad de cuotas',
    y = 'Precio',
    hue = 'Tienda',
    palette="Set2",

)

plt.title('Cantidad de cuotas por precio y tienda')
plt.show()
```

```python
# Cantidad de cuotas promedio por tienda
df_cuotas = df.groupby(by='Tienda')['Cantidad de cuotas'].mean().round(0).reset_index()
```

```python
# Resultado
print_reporte(df_cuotas, 'CANTIDAD DE CUOTAS EN PROMEDIO POR TIENDA', 'Tienda', 'Cantidad de cuotas')
```

‚≠ï **Interpretaci√≥n**  

Todas las tiendas presentan una distribuci√≥n similar en cuanto a las cantidades de cuotas de acuerdo al precio con algunas exepciones que no son relevantes para el an√°lisis.  
Por otro lado en promedio de cantidad de cuotas todas rondan las 3 cuotas.

# üî¢ Valoraci√≥n media por tienda

```python
# Agrupamos por tienda y calculamos el promedio de la calificaci√≥n
df_valoracion =  df.groupby(by='Tienda')['Calificaci√≥n'].mean().round(2).reset_index()
```

```python
# Resultado
print_reporte(df_valoracion, 'VALORACI√ìN PROMEDIO POR TIENDA', 'Tienda', 'Calificaci√≥n')
```

‚≠ï **Interpretaci√≥n**  
Tienda_1 y Tienda_4 tiene el promedio de calificaci√≥n m√°s baja.

# üèÜ Productos m√°s vendidos

```python
# Agrupamos por tienda y producto
df_productos = df.groupby('Tienda')['Producto'].value_counts().reset_index(name='Cantidad')
```

```python
# Resultado Productos m√°s vendidos

for i in range(1,5):
    tienda = f'Tienda_{i}'

    filtro = df_productos['Tienda'] == tienda

    df_aux = df_productos[filtro].head()
    
    print_reporte(df_aux, f'PRODUCTOS M√ÅS VENDIDOS EN LA TIENDA {i}', 'Tienda', 'Producto')

    print()
```

```python
# Resultado Productos menos vendidos

for i in range(1,5):
    tienda = f'Tienda_{i}'

    filtro = df_productos['Tienda'] == tienda

    df_aux = df_productos[filtro].tail()
    
    print_reporte(df_aux, f'PRODUCTOS MENOS VENDIDOS EN LA TIENDA {i}', 'Tienda', 'Producto')

    print()
```

# üè£ Valor promedio de env√≠o por tienda y rentabilidad

```python
# Agrupamos por tienda y sacamos el promedio de costo de env√≠o
df_envios = df.groupby('Tienda')['Costo de env√≠o'].mean().reset_index()
```

```python
# Mostramos los resultados
print_reporte(df_envios, 'VALOR PROMEDIO DE ENV√çO POR TIENDA', 'Tienda', 'Costo de env√≠o', True)
```

‚≠ï **Interpretaci√≥n**  
La tienda 4 es la que m√°s barato le cuesta hacer los env√≠os y la tienda 2 y 1 son las que m√°s gastar√≠an de momento en env√≠os.

```python
# Margen aproximado por tienda
df_margen = df.copy()

df_margen['Margen'] = df_margen['Precio'] - df_margen['Costo de env√≠o']
```

```python
# Agrupamos por tienda y sumamos el margen y sacamos promedio
df_margen_agg = df_margen.groupby(by='Tienda')['Margen'].agg(Promedio='mean', Suma='sum')
```

```python
# Resultado

sns.barplot(
    data = df_margen_agg,
    x = 'Promedio',
    y = df_margen_agg.index
)

plt.title('MARGEN APROXIMADO POR TIENDA')
plt.show()
```

‚≠ï **Interpretaci√≥n**  
La tienda 4 es la que menos margen aproximado da.

```python
# Distribuci√≥n geogr√°fica de las ventas

df_count = df.groupby(["lat", "lon"]).size().reset_index(name="cantidad")

fig = px.scatter_geo(
    df_count,
    lat="lat",
    lon="lon",
    size="cantidad",
    projection="equirectangular",
    scope="south america"
)

fig.show()
```
